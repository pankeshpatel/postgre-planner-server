```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
df = pd.read_csv('/Users/pankeshpatel/Desktop/bmw-material-planner-assistant/backend/total_qty.csv') 
dataframe = pd.DataFrame(df, columns=['material', 'demand_date', 'max', 'min', 'mean', 'safety stock'])
dataframe
```

```python
font = {
        'color':  'black',
        'weight': '10',
        'size': 10,
        }




# Create figure
fig = plt.figure(figsize=(16, 10))

# Plot time series
plt.plot(dataframe['demand_date'], dataframe['max'], marker='o', label='Max of Total Quatity', linewidth=2)
plt.plot(dataframe['demand_date'], dataframe['min'], marker='o', label='Min of Total Quatity', linewidth=2)
plt.plot(dataframe['demand_date'], dataframe['mean'], marker='o', label='Average of Total Quatity', linewidth=2)
plt.plot(dataframe['demand_date'], dataframe['safety stock'], marker='o', label='Safety Stock', linewidth=2)


plt.title(label = 'HealthScore Analysis of Material {} of Next 10 days'.format(dataframe['material'][0]),
         fontsize=15,
         color="black",
         pad='2.0',
         fontweight=10,
         loc="center")

plt.xlabel('Date', fontdict=font)
plt.ylabel('Quantity of Material', fontdict=font)

# # Add legend
plt.legend(loc='best')

plt.ylim(0,1000)


index = dataframe['demand_date']
max = dataframe['max']
min = dataframe['min']
mean = dataframe['mean']
safety_stock = dataframe['safety stock']

# max
for x, y in zip(index, max):
    label = y
    plt.annotate(label, (x, y),
                 xycoords="data",
                 textcoords="offset points",
                 xytext=(0, 12), ha="center")
    

# min
for x, y in zip(index, min):
    label = y
    plt.annotate(label, (x, y),
                 xycoords="data",
                 textcoords="offset points",
                 xytext=(0, 12), ha="center")
    
# mean
for x, y in zip(index, mean):
    label = y
    plt.annotate(label, (x, y),
                 xycoords="data",
                 textcoords="offset points",
                 xytext=(0, 12), ha="center")
    

# safety_stock
for x, y in zip(index, safety_stock):
    label = y
    plt.annotate(label, (x, y),
                 xycoords="data",
                 textcoords="offset points",
                 xytext=(0, 12), ha="center")


# # Auto space
plt.tight_layout()

# # Display plot
plt.show() 
```

```python
# Data Reading from CSV file
df_exception = pd.read_csv('/Users/pankeshpatel/Desktop/exceptions-combine.csv')
dataframe_exception = pd.DataFrame(df_exception, columns=[ 'matnr' , 'cdate', 'auskt'])
```

```python
#  Data cleaning, replacing NaN with '0'
dataframe_exception['auskt'] = dataframe_exception['auskt'].fillna(0)
```

```python
# Data filtering with respect to the start and end date
dataframe_exception_filtered = dataframe_exception.filter_date('cdate', '2022-02-18', '2022-04-04')

# Remove row that 'auskt' value has zero
data_filter = dataframe_exception_filtered[dataframe_exception_filtered['auskt'] > 0]

# This would display all rows of a panda dataframe
pd.set_option('display.max_rows', data_filter.shape[0]+1)
```

```python
# "count" column
exception_count = data_filter.groupby('matnr').count()
exception_count.rename(columns = {'auskt':'count'}, inplace = True)
exception_count.drop('cdate', axis =1, inplace = True)

# 'percentage'
exception_percentage = ((exception_count['count'] / len(data_filter))*100).to_frame()
exception_percentage.rename(columns = {'count':'percentage'}, inplace = True)

# combine "count" and "percentage" column into one Table
result = pd.concat([exception_count, exception_percentage], axis=1)

# reset index
result.reset_index(inplace=True)
exception_matrix = result.rename(columns = {'matnr':'material'})
exception_matrix
```

```python
import matplotlib.pyplot as plt



#list_exception_matrix_material= (list(exception_matrix['material']))
list_exception_matrix_material = ['741788607', '742065710', '743093505', '807165907', '809198305']

list_exception_matrix_percentage = list(exception_matrix['percentage'])
# list_exception_matrix_percentage

# # Create figure
fig = plt.figure(figsize=(10, 7))

bar_colors = ['red','blue','seagreen','crimson','gold']


plt.barh(list_exception_matrix_material, list_exception_matrix_percentage, color=bar_colors)

plt.title(label = 'Part Exception Matrix (Last 45 Days)',
         fontsize=15,
         color="black",
         pad='2.0',
         fontweight=10,
         loc="center")

plt.xlabel('Percentage', fontdict=font)
plt.show()


```
